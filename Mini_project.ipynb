```python
marks = {
    "tamil": 85,
    "english": 90,
    "maths": 95,
    "science": 80,
    "social": 75
}

total = 0
for subject, mark in marks.items():
    print(subject, ":", mark)
    total += mark

print("total mark", total)

if total >= 450:
    print("grade A")
elif total >= 400:
    print("grade B")
elif total >= 350:
    print("grade C")
else:
    print("grade D")
```

```python
import numpy as np
sales=np.array([1000,1500,2000,2500,3000])
total_sales=sum(sales)
average_sales=total_sales/len(sales)
highest_sale=max(sales)
lowest_sale=min(sales)
print("Monthly Sales:",sales)
print("Total Sales:",total_sales)
print("Average Sales:",average_sales)
print("Highest Sale:",highest_sale)
print("Lowest Sale:",lowest_sale)
```

```python
import pandas as pd
employee={
    "ID":[355,764,986],
    "Name":["dharshini","radha","panumathi"],
    "Department":["IT","HR","IT"],
    "Salary":[500000,35500,76500],
    "Experience":[3,7,2]
}
df=pd.DataFrame(employee)
print("Employee Dataset")
print(df)

print("\nAverage Salary")
print(df['Salary'].mean())

print("\nHighest Salary")
print(df['Salary'].max())

print("\nEmployee in HR")
print(df[df['Department']=='HR'])

print("\nEmployee with experience more than 5 years")
print(df[df['Experience']>5])

```

```python
import pandas as pd
data={
    "Region":["South","North","East","West","South","North"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet"],
    "Sales":[500000,30000,70000,2000000,588000,40000]
}
df=pd.DataFrame(data)
print(df)
```

```python

print(df.columns)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
prodect_sales=df.groupby("Product")["Sales"].sum()
print(prodect_sales)
```

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May"]
sales=[10000,20000,300000,40000,50000]
plt.plot(months,sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May"]
sales=[10000,20000,300000,40000,50000]
plt.plot(months,sales,
        color="green",
        marker="o",
        linestyle="--")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
